# Customer Churn Analysis

## Section 1: Churn Distribution

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x='Churn Label'
)

plt.title('Customer Churn Distribution')
plt.show()
```

### Insight

- 26.5% of customers have churned.
- Customer retention should be a business priority.

## Section 2: Contract Type vs Churn

```python
contract_churn = pd.crosstab(
    df['Contract'],
    df['Churn Label'],
    normalize='index'
)*100

contract_churn
```

### Visualization:

```python
contract_churn.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5)
)

plt.ylabel("Percentage")
plt.title("Contract Type vs Churn")
plt.show()
```

### Expected Insight

- Month-to-month customers exhibit significantly higher churn rates than customers on one-year or two-year contracts.

## Section 3: Tenure Analysis

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Churn Label',
    y='Tenure Months'
)

plt.show()
```

### Insight
- Customers who churn tend to have substantially lower tenure, suggesting early-stage retention is critical.

## Section 4: Monthly Charges Analysis

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Churn Label',
    y='Monthly Charges'
)

plt.show()
```

### Insight
- Customers with higher monthly charges show increased churn tendencies.

## Section 5: Payment Method Analysis

```python
payment_churn = pd.crosstab(
    df['Payment Method'],
    df['Churn Label'],
    normalize='index'
)*100

payment_churn
```

### Visualization

```python
payment_churn['Yes'].sort_values().plot(
    kind='barh',
    figsize=(8,5)
)

plt.xlabel("Churn Rate (%)")
plt.show()
```

### Insight
- Certain payment methods demonstrate considerably higher churn rates, indicating potential friction in billing processes.

## Section 6: Internet Service Analysis

```python
internet_churn = pd.crosstab(
    df['Internet Service'],
    df['Churn Label'],
    normalize='index'
)*100

internet_churn
```

```python
internet_churn['Yes'].plot(
    kind='bar',
    figsize=(8,5)
)

plt.ylabel("Churn Rate (%)")
plt.show()
```

## Section 7: Tech Support Impact

```python
tech_churn = pd.crosstab(
    df['Tech Support'],
    df['Churn Label'],
    normalize='index'
)*100

tech_churn
```

```python
tech_churn['Yes'].plot(
    kind='bar'
)

plt.ylabel("Churn Rate (%)")
plt.show()
```

### Insight
- Customers without technical support churn at a significantly higher rate.

## Section 8: Churn Reasons

```python
top_reasons = (
    df[df['Churn Label']=='Yes']
    ['Churn Reason']
    .value_counts()
    .head(10)
)

top_reasons
```

### Visualization

```python
top_reasons.plot(
    kind='barh',
    figsize=(10,6)
)

plt.title("Top Churn Reasons")
plt.show()
```

## Section 9: Revenue at Risk

```python
revenue_at_risk = (
    df.loc[
        df['Churn Label']=='Yes',
        'Monthly Charges'
    ].sum()
)

print(revenue_at_risk)
```

## Section 10: Executive Summary

## Key Findings

- Overall churn rate is 26.5%.
- Month-to-month contracts show the highest churn rate.
- Customers with lower tenure are more likely to leave.
- High monthly charges are associated with increased churn.
- Lack of technical support strongly correlates with churn.
- Certain payment methods experience higher churn.
- Revenue loss due to churn represents a significant business risk.
