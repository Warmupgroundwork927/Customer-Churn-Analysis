# Customer Churn Analysis

## Dataset Overview

- Total Customers: 7043
- Churned Customers: 1869
- Active Customers: 5174
- Churn Rate: 26.5%

```python
print(f"Total Customers: {len(df)}")
print(f"Churned Customers: {(df['Churn Label']=='Yes').sum()}")
print(f"Active Customers: {(df['Churn Label']=='No').sum()}")
```

## Section 1: Churn Distribution

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x='Churn Label'
)

plt.title('Customer Churn Distribution')
plt.show()
```

### Insight

- 26.5% of customers have churned.
- Customer retention should be a business priority.

## Section 2: Contract Type vs Churn

```python
contract_churn = pd.crosstab(
    df['Contract'],
    df['Churn Label'],
    normalize='index'
)*100

contract_churn
```

```python
contract_churn['Yes'].sort_values(ascending=False)
```

### Visualization:

```python
contract_churn['Yes'].sort_values(ascending=False).plot(
    kind='bar',
    figsize=(8,5)
)

plt.ylabel('Churn Rate (%)')
plt.title('Churn Rate by Contract Type')
plt.show()
```

### Insight

- Month-to-month customers exhibit significantly higher churn rates than customers on one-year or two-year contracts.

## Section 3: Tenure Analysis

```python
tenure_churn = (
    df.groupby('Churn Label')['Tenure Months']
    .mean()
    .sort_values()
)

tenure_churn
```

```python
tenure_churn.plot(
    kind='bar',
    figsize=(6,4)
)

plt.ylabel('Average Tenure (Months)')
plt.title('Average Tenure by Churn Status')
plt.xticks(rotation=0)
plt.show()
```

### Insight

- Churned customers have significantly lower average tenure than retained customers.
- Customer retention efforts should focus on customers within their first year of service.

## Section 4: Monthly Charges Analysis

```python
monthly_charges = (
    df.groupby('Churn Label')['Monthly Charges']
    .mean()
    .sort_values()
)

monthly_charges
```

```python
monthly_charges.plot(
    kind='bar',
    figsize=(6,4)
)

plt.ylabel('Average Monthly Charges ($)')
plt.title('Average Monthly Charges by Churn Status')
plt.xticks(rotation=0)
plt.show()
```

### Insight

- Churned customers pay higher average monthly charges than retained customers.
- Pricing and perceived value may be contributing factors to customer attrition.

## Section 10: Executive Summary

## Key Findings

1. The overall customer churn rate is 26.5%, with 1,869 out of 7,043 customers leaving the service.

2. Customers on month-to-month contracts exhibit the highest churn rate at 42.7%, compared to 11.3% for one-year contracts and only 2.8% for two-year contracts.

3. Customers using Electronic Check as their payment method show the highest churn rate at 45.3%, nearly three times higher than customers using automatic payment methods.

4. Customer support quality appears to be a major churn driver. The most frequently reported churn reason is "Attitude of Support Person" (192 customers).

5. Competitive pressure significantly impacts retention. Competitor-related reasons account for several of the top churn causes, including:
   - Higher download speeds (189 customers)
   - More data offerings (162 customers)
   - Better offers (140 customers)

6. Service quality and customer experience issues contribute heavily to churn, with complaints related to service attitude, network reliability, and product dissatisfaction appearing among the top reported reasons.

7. Long-term contracts and automated payment methods are strongly associated with improved customer retention.

# Business Recommendations

1. Increase adoption of annual and two-year contracts through discounts and loyalty incentives.

2. Reduce churn among Electronic Check users by promoting automatic payment options with rewards or cashback incentives.

3. Strengthen customer support training programs to address recurring complaints regarding support staff interactions.

4. Improve competitiveness by reviewing internet plans, download speeds, and pricing structures.

5. Launch targeted retention campaigns for customers within their first year of service, as early churn contributes significantly to customer loss.

6. Monitor customers reporting service dissatisfaction and proactively offer resolution programs before cancellation.
